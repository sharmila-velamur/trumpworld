```python
from operator import itemgetter
import json
```

```python
trump_raw = open('trump-world-d3.json', 'r')
trump_object= json.loads(trump_raw.read())
```

```python
trump_object['nodes']
```

```python
results = [x for x in trump_object['nodes'] if 'pr' in x]
```

```python
sorted_pr  = sorted(results, key = itemgetter('pr'), reverse = True)
```

```python
sorted_pr_json = sorted_pr[:100]
```

```python
json.dumps(sorted_pr, sort_keys = True)
```

```python
with open('sorted_pr_trump.json', 'w') as outfile:
    json.dump(sorted_pr_json, outfile)
```

```python
trump_object['links']
```

```python
for p in sorted_pr_json[:20]:
    print(p["name"], p["pr"])
```

```python
###Degree Centrality
results = [x for x in trump_object['nodes'] if 'dcn' in x]

#sort
sorted_dcn  = sorted(results, key = itemgetter('dcn'), reverse = True)

for p in sorted_dcn[:20]:
    print(p["name"], p["dcn"])
```
